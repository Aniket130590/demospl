```python
def GetCorrelations(x,y,df):
    import pandas as pd
    ds=df.apply(lambda x : pd.factorize(x)[0])
    #target = 'Number'
    k = x #number of variables for heatmap
    cols = ds.corr().nlargest(k, y)[y].index
    cm = ds[cols].corr().abs()
    return cm
```

```python
# fuzz is used to compare TWO strings
from fuzzywuzzy import fuzz

# process is used to compare a string to MULTIPLE other strings
from fuzzywuzzy import process
```

```python
def get_ratio(row):
    name = row['name']
    return fuzz.token_sort_ratio(name, "ABBAS")

def get_ratio2(row):
    name = row['addresses']
    return fuzz.token_sort_ratio(name, "SY")

df1 = df[df.apply(get_ratio, axis=1) > 50]
df2 = df1[df1.apply(get_ratio2, axis=1) > 10]
```
